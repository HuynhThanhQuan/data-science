Gini Impurity and Entropy

$Gini(D) = 1 - \sum_{i=1}^{k}{p_{i}^{2}}$

Which:
- $p_{i}$ is the probability/frequency of class i in dataset

```python
def cal_gini_impurity_from_dict(d_data):
    n_classes = len(d_data.keys())
    total = sum(d_data.values())
    p_c = sum([(i/total)**2 for i in d_data.values()])
    gini_imp = 1 - p_c
    return gini_imp, d_data

def cal_gini_impurity_from_list(data):
    d_data = {}
    for i in data:
        d_data[i] = d_data.get(i, 0) + 1
    return cal_gini_impurity_from_dict(d_data)
```

## Weighted Gini Impurity

Weighted Gini Impurity = $\frac{n_{1}}{n} \times Gini_{1} + \frac{n_{2}}{n} \times Gini_{2}$

We going to calculate the Weighted Gini Impurity of Gender attribute based on the split at unique values

```python
feat_gini = []
```

```python
n = len(tmp_df)

weight_gini = 0

record_data =[]
for v in df[sel_feat].unique():
    data = tmp_df[tmp_df[sel_feat]==v]['fraud_label']
    n_v = len(data)
    v_gini, d_data = cal_gini_impurity_from_list(data)
    n_good = d_data.get('Good', 0)
    n_bad = d_data.get('Bad', 0)
    
    weight_gini += n_v / n * v_gini
    print(f'Feature {sel_feat} split at {v} has Gini \t {v_gini}')
    record_data.append((sel_feat, v, n_good, n_bad, n_v, v_gini))

print()
print(f'Weighted Gini Impurity of {sel_feat} is \t {weight_gini}')
gini_gain =origin_gini - weight_gini
print(f'Gini Gain {gini_gain}')
```

```python
feat_gini.append((sel_feat, weight_gini, gini_gain))
```

```python
att_df = pd.DataFrame(record_data, columns=['Feature', 'Split', 'n-good', 'n-bad', 'total', 'Gini'])
att_df
```

```python
feat_report = pd.DataFrame(feat_gini,columns=['Feature', 'Weighted Gini Impurity', 'Gini Gain'])
feat_report
```

## Calculate Gini Impurity of all Categorical Features

So we will have 2 report dataframe

- 1 for Gini Impurity of each unique value of each feature
- 1 for Weighted Gini Impurity of each Feature

```python
weighted_gini = []
gini_impurity =[]
n = len(tmp_df)
```

```python
for sel_feat in cat_cols:
    tmp_df = pd.concat([df[sel_feat], label],axis=1)
    weighted_gini_imp = 0
    for v in df[sel_feat].unique():
        data = tmp_df[tmp_df[sel_feat]==v]['fraud_label']
        v_n = len(data)
        v_gini, d_data = cal_gini_impurity_from_list(data)
        n_good = d_data.get('Good', 0)
        n_bad = d_data.get('Bad', 0)
        weighted_gini_imp += v_n / n * v_gini
        gini_impurity.append((sel_feat, v, n_good, n_bad, v_n, v_gini))
    gini_gain = origin_gini - weighted_gini_imp
    weighted_gini.append((sel_feat, weighted_gini_imp, gini_gain))
```

```python
attribute_df = pd.DataFrame(gini_impurity, columns=['Feature', 'Split', 'n-good', 'n-bad', 'total', 'Gini'])
feature_report = pd.DataFrame(weighted_gini,columns=['Feature', 'Weighted Gini Impurity', 'Gini Gain'])
```
