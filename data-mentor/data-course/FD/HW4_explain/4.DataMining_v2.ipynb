## 1. Transaction History & Payment & Promotion

### Business Logic

Payment và Transaction khác nhau như thế nào?

> Khi bạn mua một chiếc điện thoại trực tuyến, việc bạn thực hiện thanh toán bằng thẻ tín dụng hoặc chuyển khoản ngân hàng là một payment (thanh toán). Trong khi đó, cả quá trình mua hàng này, từ khi bạn chọn sản phẩm, thực hiện thanh toán, và nhận sản phẩm là một transaction (giao dịch).

Transaction (giao dịch): Là quá trình mua bán chiếc điện thoại, từ khi bạn chọn chiếc điện thoại và quyết định mua đến khi bạn thực hiện thanh toán và nhận chiếc điện thoại. Giao dịch này bao gồm tất cả các hoạt động liên quan đến việc mua hàng, bao gồm cả việc chọn sản phẩm, thực hiện thanh toán và nhận sản phẩm.

Payment (thanh toán): Là phần trong giao dịch mà bạn thực hiện việc thanh toán bằng thẻ tín dụng. Khi bạn chọn chiếc điện thoại và quyết định mua, bạn tiến hành thanh toán bằng thẻ tín dụng, và thông tin thanh toán được gửi đi để xử lý thanh toán. Việc này là một phần của giao dịch, và nó là một payment (thanh toán) trong quá trình mua hàng.

Promotion Campaign là những chiến dịch khuyến mãi nhằm tăng sự tương tác và thu hút khách hàng sử dụng sản phẩm của họ

Với dạng feature này, chúng ta sẽ có kỹ thuật binning để nhóm các continuous number lại thành group

```python
def group_rating(x):
    if x >= 4.5:
        return 5
    elif x >= 3.5:
        return 4
    elif x >= 2.5:
        return 3
    elif x >= 1.5:
        return 2
    elif x >= 0:
        return 1
    else:
        return None

df['rating_group'] = df['rating'].apply(lambda x: group_rating(x))
```

```python
df['rating_group'].value_counts().sort_index(ascending=True).plot(kind='barh')
```

- Đếm số lượng transaction trên mỗi user trong 3,6,9,12,24 tháng gần nhất

```python
q = """
SELECT 
    customer_id,
    COUNT(CASE WHEN transaction_datetime >= DATE('2023-01-01', '-3 months') THEN 1 END) AS transaction_count_3_months,
    COUNT(CASE WHEN transaction_datetime >= DATE('2023-01-01', '-6 months') THEN 1 END) AS transaction_count_6_months,
    COUNT(CASE WHEN transaction_datetime >= DATE('2023-01-01', '-9 months') THEN 1 END) AS transaction_count_9_months,
    COUNT(CASE WHEN transaction_datetime >= DATE('2023-01-01', '-12 months') THEN 1 END) AS transaction_count_12_months,
    COUNT(CASE WHEN transaction_datetime >= DATE('2023-01-01', '-24 months') THEN 1 END) AS transaction_count_24_months
FROM 
    df
GROUP BY 
    customer_id HAVING customer_id not NULL;

"""
ps.sqldf(q)
```

- Tổng tiền transaction trên mỗi user trong 3,6,9,12,24 tháng gần nhất

```python
q = """
SELECT 
    customer_id,
    SUM(CASE WHEN transaction_datetime >= DATE('2023-01-01', '-3 months') THEN txn_total_amount ELSE 0 END) AS transaction_count_3_months,
    SUM(CASE WHEN transaction_datetime >= DATE('2023-01-01', '-6 months') THEN txn_total_amount ELSE 0 END) AS transaction_count_6_months,
    SUM(CASE WHEN transaction_datetime >= DATE('2023-01-01', '-9 months') THEN txn_total_amount ELSE 0 END) AS transaction_count_9_months,
    SUM(CASE WHEN transaction_datetime >= DATE('2023-01-01', '-12 months') THEN txn_total_amount ELSE 0 END) AS transaction_count_12_months,
    SUM(CASE WHEN transaction_datetime >= DATE('2023-01-01', '-24 months') THEN txn_total_amount ELSE 0 END) AS transaction_count_24_months
FROM 
    df
GROUP BY 
    customer_id HAVING customer_id not NULL;

"""
ps.sqldf(q)
```

- Đếm số lượng transaction mà khách hàng hủy

```python
q = """
SELECT 
    customer_id,
    COUNT(CASE WHEN transaction_status = 'Đã hủy' AND transaction_datetime >= DATE('2023-01-01', '-3 months') THEN 1 END) AS canceled_count_3_months,
    COUNT(CASE WHEN transaction_status = 'Đã hủy' AND transaction_datetime >= DATE('2023-01-01', '-6 months') THEN 1 END) AS canceled_count_6_months,
    COUNT(CASE WHEN transaction_status = 'Đã hủy' AND transaction_datetime >= DATE('2023-01-01', '-9 months') THEN 1 END) AS canceled_count_9_months,
    COUNT(CASE WHEN transaction_status = 'Đã hủy' AND transaction_datetime >= DATE('2023-01-01', '-12 months') THEN 1 END) AS canceled_count_12_months,
    COUNT(CASE WHEN transaction_status = 'Đã hủy' AND transaction_datetime >= DATE('2023-01-01', '-24 months') THEN 1 END) AS canceled_count_24_months
FROM 
    df
GROUP BY 
    customer_id HAVING customer_id not NULL;
"""
ps.sqldf(q)
```

 - 1 ví dụ sử dụng partition by, để tính total_amount mỗi tháng của mỗi khách hàng

```python
q = """
SELECT * FROM
(SELECT
    customer_id,
    strftime('%Y-%m', transaction_datetime) AS month,
    SUM(total_amount) OVER (PARTITION BY customer_id, strftime('%Y-%m', transaction_datetime)) AS total_amount_per_month
FROM
    df
ORDER BY
    customer_id, month)
WHERE customer_id is not NULL and month is not NULL
"""
ps.sqldf(q).drop_duplicates()
```
