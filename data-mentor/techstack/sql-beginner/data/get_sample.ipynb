# DEMOGRAPHIC

```python
df = pd.read_csv('demographic.csv',index_col=0)
```

```python
cols = df.columns.tolist()
cols.remove('customer_id')
```

```python
for col in cols:
    df.loc[df.sample(frac=0.1).index, col] = None
```

```python
df = df[:100]
```

```python
df.to_csv('sql_demo.csv',index=False)
```

```python
df.columns.tolist()
```

```python
pd.read_csv('sql_demo.csv')
```

# MEMBER

```python
df = pd.read_csv('member.csv',index_col=0)
df = df.drop(columns=['member_lifespan'])
```

```python
cols = df.columns.tolist()
cols.remove('customer_id')
```

```python
for col in cols:
    df.loc[df.sample(frac=0.05).index, col] = None
```

```python
df = df[:150]
```

```python
df.to_csv('sql_member.csv',index=False)
```

```python
df.columns.tolist()
```

```python
pd.read_csv('sql_member.csv')
```
